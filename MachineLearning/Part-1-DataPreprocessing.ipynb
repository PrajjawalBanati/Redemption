**Feature Scaling**

- Always applied to columns. Not applied across the row.
- Two types of Feature scaling techiniqies.
  - **Normalization**
    - It refers to following formula.<br/>
$$f(X_{'}) = X - X_{min} / X_{max} - X_{min}$$
    - For any value $X$ we get the $X_{min}$ and $X_{max}$ present in the column.
    - Its value always remains between $[0,1]$.
   
  - **Standardization**
    - Here we calculate Standardization using the following formula.<br/>
$$f(X_{'}) = X - Î¼ / \sigma $$
      - $\mu$ - average of the data present in the column.
      - $\sigma$ - standard deviation constant.
    - Its values always remains between $[-3 ; +3]$.

    - Illustration: Suppose we have 3 persons and their total income and their age is as follows.

<center>

| Person | Income      | Age |
| :------| :---------- |:----|
| A | 70,000  | 45 yrs |
| B | 60,000 | 44 yrs |
| C | 52,000 | 40 yrs |

</center>

  - Now we have to calculate that if we were need to group person B with any of the two persons, according to his age and salary, who will be the best person? Person A or C.

  - We are trying to compare two complete unrelated values, one is salary which is ranging in thousands and the ages in just in numbers. So these larger values could over power the decison by thinking that person B is much more related to C as there is only a difference of 8,000 in the income and thus your age parameter gets ignored.

  - So here comes the normalisation feature which will make each value upto a same range for both columns so that we can exactly check and make a balanced decision.

<center>

| Person | Income      | Age |
| :------| :---------- |:----|
| A | 1  | 1 |
| B | 0.444 | 0.8 |
| C | 0 | 0 |

</center> 

So now you can see that Person B is very much near to Person A. Age paramter is so close to person A however income parameter is very close to middle of A and C. So here you can see person B could be grouped with person A and that is the best decision we could make, just because we had the Normalization power.
